```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# Crear un circuito con 1 qubit
qc = QuantumCircuit(1, 1)
qc.h(0)  # Aplicar Hadamard para obtener |+⟩
qc.measure(0, 0)

# Mostrar el circuito
print(qc.draw())

# Simular
simulator = AerSimulator()
result = simulator.run(qc, shots=500).result()
counts = result.get_counts()

# Calcular probabilidades
total_shots = 500
prob_0 = counts.get('0', 0) / total_shots
prob_1 = counts.get('1', 0) / total_shots

# Mostrar resultados
print(counts)
print(f"Probabilidad |0⟩: {prob_0:.3f}")
print(f"Probabilidad |1⟩: {prob_1:.3f}")

# Graficar probabilidades
plt.bar(['|0⟩', '|1⟩'], [prob_0, prob_1], color=['blue', 'orange'])
plt.ylabel('Probabilidad')
plt.title('Probabilidades de medición')
plt.ylim(0, 1)
plt.show()
```
