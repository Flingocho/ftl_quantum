```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# Create a circuit for 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Apply Hadamard to the first qubit to create superposition
qc.h(0)

# Apply CNOT to entangle the qubits
qc.cx(0, 1)

# Measure both qubits
qc.measure([0, 1], [0, 1])

# Display the circuit
print(qc.draw())

# Simulate the circuit
simulator = AerSimulator()
result = simulator.run(qc, shots=500).result()
counts = result.get_counts()

# Display the results
print("\nMeasurement counts:")
print(counts)
plt.figure(figsize=(8, 6))
plt.bar(counts.keys(), counts.values(), color=['blue', 'orange'])
plt.xlabel('Quantum States')
plt.ylabel('Counts')
plt.title('Measurement Results for Bell State (|00⟩ + |11⟩)/√2')
plt.ylim(0, max(counts.values()) * 1.1)
plt.show()
```
