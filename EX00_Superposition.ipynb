```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# Create a circuit with 1 qubit
qc = QuantumCircuit(1, 1)
qc.h(0)  # Apply Hadamard to get |+⟩
qc.measure(0, 0)

# Show the circuit
print(qc.draw())

# Simulate
simulator = AerSimulator()
result = simulator.run(qc, shots=500).result()
counts = result.get_counts()

# Calculate probabilities
total_shots = 500
prob_0 = counts.get('0', 0) / total_shots
prob_1 = counts.get('1', 0) / total_shots

# Show results
print(counts)
print(f"Probability |0⟩: {prob_0:.3f}")
print(f"Probability |1⟩: {prob_1:.3f}")

# Plot probabilities
plt.bar(['|0⟩', '|1⟩'], [prob_0, prob_1], color=['blue', 'orange'])
plt.ylabel('Probability')
plt.title('Measurement Probabilities')
plt.ylim(0, 1)
plt.show()
```
